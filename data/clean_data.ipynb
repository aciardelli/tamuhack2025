```python
import pandas as pd
```

```python
df = pd.read_csv('toyota_vehicles_with_specs.csv')
df.head()
```

```python
print(df.columns)
```

```python
df['shown_price'] = df['shown_price'].str.extract(r'\$([0-9,]+)').replace(',','', regex=True).astype(float)
```

```python
df.head()
```

```python
df['header'] = df['header'].str.extract(r'(\d+(?:,\d+)?(?:\.\d+)?)').replace(',','', regex=True).astype(float)
df.head()
```

```python
df = df.rename(columns={'header': 'msrp'})
df.head()
```

```python
df.to_csv('toyta_final.csv')
```
