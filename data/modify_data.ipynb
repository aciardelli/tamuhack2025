```python
def calculate_horsepower(df):
    """
    Calculate more accurate horsepower estimates based on vehicle attributes.
    """
    hp = pd.Series(index=df.index, dtype=int)

    # Gasoline vehicles: Estimate horsepower based on displacement, cylinders, and vehicle type
    gas_mask = df['is_gas'] == 1
    hp[gas_mask] = (
        df.loc[gas_mask, 'displacement'] * 50 +  # Higher factor for displacement
        df.loc[gas_mask, 'cylinders'] * 15 +    # Moderate adjustment for cylinder count
        np.where(
            df.loc[gas_mask, 'class'].str.contains('SUV', case=False, na=False), 20, 0
        )  # SUVs get a bonus
    ).astype(int)

    # Electric vehicles: Estimate horsepower using city_mpg (as a proxy for motor power)
    electric_mask = df['is_electric'] == 1
    hp[electric_mask] = (
        80 + df.loc[electric_mask, 'city_mpg'] * 0.6  # Higher base for EVs
    ).astype(int)

    # Fuel cell vehicles: Assign an approximate horsepower
    fuel_cell_mask = df['is_fuel_cell'] == 1
    hp[fuel_cell_mask] = 182  # Example: Similar to Toyota Mirai's power output

    # Edge cases: Assign a default value for undefined or hybrid vehicles
    undefined_mask = ~(gas_mask | electric_mask | fuel_cell_mask)
    hp[undefined_mask] = 150

    return hp

# Apply the refined function to the dataframe
df['horsepower'] = calculate_horsepower(df)
```

```python
def calculate_financials(df):
    """
    Adds lease_price_per_month and finance_price_per_month columns to a DataFrame.

    Parameters:
        df (pd.DataFrame): DataFrame containing car data with 'shown_price' or 'msrp'.

    Returns:
        pd.DataFrame: Updated DataFrame with new financial columns.
    """
    def calculate_lease_price(price):
        residual_value = 0.55 * price  # 55% residual value at the end of the lease
        money_factor = 0.002  # Approx. 5% APR divided by 2400
        lease_term = 36  # Lease term in months
        lease_price = ((price - residual_value) / lease_term) + (price + residual_value) * money_factor
        return round(lease_price, 2)

    def calculate_finance_price(price):
        apr = 0.05  # 5% annual percentage rate
        finance_term = 60  # Finance term in months
        monthly_interest_rate = apr / 12
        finance_price = (price * monthly_interest_rate) / (1 - (1 + monthly_interest_rate) ** -finance_term)
        return round(finance_price, 2)

    # Use 'shown_price' if available, otherwise fallback to 'msrp'
    df["price_to_use"] = df[["shown_price", "msrp"]].max(axis=1)

    # Calculate lease and finance prices
    df["lease_price_per_month"] = df["price_to_use"].apply(calculate_lease_price)
    df["finance_price_per_month"] = df["price_to_use"].apply(calculate_finance_price)

    # Drop the temporary 'price_to_use' column
    df.drop(columns=["price_to_use"], inplace=True)

    return df

df = calculate_financials(df)
df['lease_term'] = 36
df['finance_term'] = 60
df.to_csv('toyota_lease.csv')
```
